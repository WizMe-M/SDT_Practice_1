```python
# библиотеки

import time
from IPython.display import clear_output
from math import factorial, sqrt, log, exp
```

```python
# вспомогательные методы
def clear(delay = 1):
    time.sleep(delay)
    clear_output(wait = True)

def show(text, delay = 1):
    clear(delay)
    print(text)
    
def await_any_input():
    print("\nВведите что угодно")
    return input()

def await_exit():
    print("\nВведите что угодно (0, чтобы вернуться)")
    if input() == "0": 
        return True

def get_int_input(header, var_name):
    while True:
        show(header)
        try:
            num = int(input(f"{var_name} = "))
            return num
        except:
            print(f"'{var_name}' не было целым числом")
        
def get_float_input(header, var_name):
    while True:
        show(header)
        try:
            num = float(input(f"{var_name} = ").replace(',', '.'))
            return num
        except:
            print(f"'{var_name}' не было вещественным числом")
```

```python
# Калькулятор (первая функции программы)

def choose_calc_func():    
    description = """
Выберите функцию кальулятора:
1)  Сумма
2)  Вычитание
3)  Умножение
4)  Деление
5)  Деление без остатка
6)  Остаток от деления
7)  Возведение в степень
8)  Факториал
9)  Корень числа
10) Логарифм числа по основанию
11) Экспонента
12) Периметр параллелепипеда
13) Площадь параллелепипеда

0) Выход
"""
    while True:
        show(description)
        try:
            option = int(input("Опция №"))
        except:
            print("Убедитесь, что ввели опцию верно")
            continue
        
        if option == 1:
            calc_sum()
        elif option == 2:
            calc_sub()
        elif option == 3:
            calc_mult()
        elif option == 4:
            calc_div()
        elif option == 5:
            calc_div_comp()
        elif option == 6:
            calc_div_rem()
        elif option == 7:
            calc_pow()
        elif option == 8:
            calc_fact()
        elif option == 9:
            calc_sqrt()
        elif option == 10:
            calc_log()
        elif option == 11:
            calc_exp()
        elif option == 12:
            calc_per()
        elif option == 13:
            calc_square()
        elif option == 0:
            break
        else:
            print("Неправильный номер опции")
        await_any_input()
        
def calc_sum():
    header = "Сумма чисел a и b"
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    show(f"Сумма a ({a}) и b ({b}) = {a + b}")
    
def calc_sub():
    header = "Вычитание из числа a числа b"
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    show(f"Вычитание из числа a ({a}) числа b ({b}) = {a - b}")
        
def calc_mult():
    header = "Умножение числа a на число b"
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    show(f"Умножение числа a ({a}) на число b ({b}) = {a * b}")
    
def calc_div():
    header = "Деление числа а на число b"
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    while b == 0:
        print("Второе число (b) не может быть равно 0!")
        b = get_float_input(header, 'b')
    show(f"Деление числа a ({a}) на число b ({b}) = {a / b}")

def calc_div_comp():
    header = "Деление числа а на число b нацело"
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    while b == 0:
        print("Второе число (b) не может быть равно 0!")
        b = get_float_input(header, 'b')
    show(f"Деление числа a ({a}) на число b ({b}) нацело = {a // b}")
    
def calc_div_rem():
    header = "Остаток от деления числа a на число b"
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    while b == 0:
        print("Второе число (b) не может быть равно 0!")
        b = get_float_input(header, 'b')
    show(f"Остаток от деления числа a ({a}) на число b ({b}) = {a % b}")

def calc_pow():
    header = "Возведение числа a в степень числа b"
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    show(f"Возведение числа a ({a}) в степень числа b ({b}) = {a ** b}")
    
def calc_fact():
    header = "Факториал числа a"
    a = get_int_input(header, 'a')
    show(f"Факториал числа a ({a}) = {factorial(a)}")
    
def calc_sqrt():
    header = "Корень числа a"
    a = get_float_input(header, 'a')
    show(f"Корень числа a ({a}) = {sqrt(a)}")

def calc_log():
    header = "Логарифм числа а по основанию b"
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    show(f"Логарифм числа a ({a}) по основанию b ({b}) = {log(a, b)}")
    
def calc_exp():
    header = "Экспонента (число Эйлера в степени a)"
    a = get_float_input(header, 'a')
    show(f"Экспонента = {exp(a)}")

def calc_per():
    header = "Периметр параллелепипеда; Формула: P = 4(a+b+c)"
    perimeter = lambda a, b, c: 4 * (a + b + c)
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    c = get_float_input(header, 'c')
    show(f"Периметр параллелепипеда со сторонами a ({a}), b ({b}) и c ({c}) = {perimeter(a, b, c)}")


def calc_square():
    header = "Площадь параллелепипеда; Формула: S = 2(ab+bc+ac)"
    square = lambda a, b, c: 2 * (a * b + b * c + a * c)
    a = get_float_input(header, 'a')
    b = get_float_input(header, 'b')
    c = get_float_input(header, 'c')
    show(f"Площадь параллелепипеда со сторонами a ({a}), b ({b}) и c ({c}) = {square(a, b, c)}")
```

```python
# Подсчет в строке (вторая функция программы)
def count_in_string():
    description = """
Подсчёт символов в строке (количество запятых, пробелов и общее количество символов)
Введите текст, который необходимо проанализировать
"""
    is_exit = False
    while not is_exit:
        show(description)
        text = str(input("Текст: "))
        analyzes = analyze(text)
        for key in analyzes.keys():
            print(f"{key} - {analyzes[key]}")
        is_exit = await_exit()        
    
# Возвращает словарь символов и их количества в анализируемой строке
def analyze(text):
    space = 'Пробелы'
    comma = 'Запятые'
    total = 'Всего символов'
    dictionary = {space: 0, comma: 0, total:0}
    for c in text:
        dictionary[total] += 1
        if c == " ":
            dictionary[space] += 1
        elif c == ",":
            dictionary[comma] += 1
    return dictionary
```

```python
# Матрица (третья функция программы)
def matrix():
    header = """
Создание и заполнение матрицы (слева направо сверху вниз)
Необходимо ввести 4 параметра:
1) 'columns' - количество столбцов матрицы (не может быть вещественным, меньше или равным 0)
2) 'rows' - количество строк (не может быть вещественным, меньше или равным 0)
3) 'first' - значение первого элемента матрицы
4) 'step' - значение, на которое будет изменяться следующий элемент матрицы
"""
    is_exit = False
    while not is_exit:
        columns = get_matrix_parameter(header, 'columns')
        rows = get_matrix_parameter(header, 'rows')
        first = get_float_input(header, 'first')
        step = get_float_input(header, 'step')  
        show(header)
        for i in range(rows):
            for j in range(columns):
                print(round(first, 3), end=' ')
                first += step
            print()
        is_exit = await_exit()
    
# Получает натуральное число, которое может означать количество столбцов или строк матрицы
def get_matrix_parameter(header, parameter_name):
    while True:
        parameter = get_int_input(header, parameter_name)
        if parameter > 0:
            return parameter
    
```

```python
# Вызовы методов и главное меню
while True:
    description = """
Добро пожаловать!
Введите число, для вызова соответствующей функциональности:
1) Калькулятор
2) Подсчет символов в строке
3) Вывод матрицы

0) Выход из приложения"""
    show(description)
    try:
        option = int(input("Опция №"))
    except:
        print("Убедитесь, что ввели опцию верно")
        continue
    
    if option == 1:
        choose_calc_func()
    elif option == 2:
        count_in_string()
    elif option == 3:
        matrix()
    elif option == 0:
        show("Вы вышли из приложения")
        break
    else:
        print("Неправильный номер опции")
```
